```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
#Math course
student_mat = pd.read_csv('/content/student-mat.csv',delimiter=';')
```

```python
#Math course in Gabriel Pereira
student_mat_gp = student_mat[student_mat['school']=='GP']
student_mat_gp.head()
```

```python
student_mat_gp.describe()
```

```python
student_mat_gp.info()
```

```python
#Total and average grade for the three quarters
total = student_mat_gp.loc[:,['G1','G2','G3']].sum(axis=1)
student_mat_gp['Avg_G'] = total/3

student_mat_gp.head()
```

```python
#Reshape DataFrame to a long format
student_mat_gp_l = student_mat_gp.melt(id_vars=['age','traveltime','studytime','failures','goout','Dalc','Walc','freetime','health','absences'],
                                       value_vars=['G1','G2','G3','Avg_G'],
                                       var_name='Period',
                                       value_name='Grade')
student_mat_gp_l.head()
```
